```python
def get_seconds(time):
    #print('Time in hh:mm:ss:', time_str)
    # split in hh, mm, ss
    hh, mm, ss = time.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
```

```python
avg_response_sec=[]
for t in df['Average Response Time']:
    avg_response_sec.append(get_seconds(t))
df['avg_response_sec'] = avg_response_sec
```

```python
df['Weekly_Report']=pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df['Week_Converted']=df['Weekly_Report'].dt.isocalendar().week
```

```python
df.head()
```

```python
df.groupby(['Agent Name','Week_Converted'])['avg_response_sec'].mean()
```
